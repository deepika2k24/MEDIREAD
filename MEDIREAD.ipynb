### ğŸ§  **Project Description**
**MediRead** is an **AI-powered system** designed to decode and digitize **handwritten medical prescriptions**.  
It combines **Convolutional Neural Networks (CNNs)** and **Optical Character Recognition (OCR)** to extract and interpret text from scanned or photographed prescriptions.  
The system integrates both a **custom-trained deep learning model** and the **OCR.Space API** for enhanced accuracy and reliability.

```python
# =======================================
# ğŸ§© 1. Setup & Installation
# =======================================
!apt-get install tesseract-ocr -y
!pip install tensorflow opencv-python pillow pytesseract scikit-learn requests matplotlib

import os
import cv2
import numpy as np
import tensorflow as tf
import pytesseract
import requests
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, Flatten, Dense,
                                     Dropout, BatchNormalization, Input)
from tensorflow.keras.preprocessing.image import img_to_array
from google.colab import files
import re
import json
import base64

print("âœ… All libraries imported successfully!")

# =======================================
# ğŸ“ 2. Enhanced Data Loading with Error Handling
# =======================================
def load_images_from_directory(directory_path, img_size=(64, 64)):
    """Safely load images with proper error handling"""
    images = []
    labels = []
    failed_files = []

    if not os.path.exists(directory_path):
        print(f"âŒ Directory {directory_path} does not exist!")
        return np.array([]), np.array([])

    for folder in os.listdir(directory_path):
        folder_path = os.path.join(directory_path, folder)
        if not os.path.isdir(folder_path):
            continue

        print(f"ğŸ“ Processing folder: {folder}")

        for img_name in os.listdir(folder_path):
            img_path = os.path.join(folder_path, img_name)

            # Skip non-image files
            if not img_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                continue

            try:
                # Read image
                img = cv2.imread(img_path)
                if img is None:
                    failed_files.append(f"Could not read: {img_path}")
                    continue

                # Convert to grayscale
                if len(img.shape) == 3:
                    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                else:
                    gray = img

                # Resize with validation
                if gray.size == 0:
                    failed_files.append(f"Empty image: {img_path}")
                    continue

                resized = cv2.resize(gray, img_size)
                normalized = resized / 255.0

                images.append(normalized)
                labels.append(folder)

            except Exception as e:
                failed_files.append(f"{img_path}: {str(e)}")
                continue

    if failed_files:
        print(f"âš ï¸ Failed to load {len(failed_files)} files")

    return np.array(images), np.array(labels), failed_files

# =======================================
# ğŸ–¼ï¸ 3. Upload and Process Dataset
# =======================================
print("ğŸ“¤ Please upload your dataset zip file:")
uploaded = files.upload()

# Extract dataset
import zipfile
for fn in uploaded.keys():
    with zipfile.ZipFile(fn, 'r') as zip_ref:
        zip_ref.extractall("prescription_data")

DATA_DIR = "prescription_data"
IMG_SIZE = (64, 64)

# Load images
X, y, failed_files = load_images_from_directory(DATA_DIR, IMG_SIZE)

if len(X) == 0:
    print("âŒ No images loaded! Creating sample data for demonstration...")
    # Create sample data for demonstration
    X = np.random.rand(50, IMG_SIZE[0], IMG_SIZE[1])
    y = np.array(['prescription'] * 50)
else:
    print(f"âœ… Successfully loaded {len(X)} images from {len(np.unique(y))} classes")

# Reshape for CNN
X = X.reshape(-1, IMG_SIZE[0], IMG_SIZE[1], 1)

# =======================================
# ğŸ”¢ 4. Data Preparation
# =======================================
# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
num_classes = len(np.unique(y_encoded))

print(f"ğŸ“Š Dataset Info:")
print(f"   - Total samples: {len(X)}")
print(f"   - Number of classes: {num_classes}")
print(f"   - Class names: {list(label_encoder.classes_)}")

# Convert to categorical
if num_classes > 1:
    y_categorical = to_categorical(y_encoded, num_classes=num_classes)
else:
    # For single class, use binary approach
    y_categorical = np.ones((len(y_encoded), 1))

# Split data
if len(X) > 1:
    X_train, X_val, y_train, y_val = train_test_split(
        X, y_categorical, test_size=0.2, random_state=42, stratify=y_encoded
    )
else:
    X_train, y_train = X, y_categorical
    X_val, y_val = X, y_categorical

print(f"âœ… Training set: {len(X_train)} samples")
print(f"âœ… Validation set: {len(X_val)} samples")

# =======================================
# ğŸ§  5. Fixed CNN Model Architecture
# =======================================
def create_prescription_model(input_shape, num_classes):
    model = Sequential()

    model.add(Input(shape=input_shape))

    # First Conv Block
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.25))

    # Second Conv Block
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.25))

    # Third Conv Block
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.25))

    # Classifier
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    # Output layer - PROPERLY CONFIGURED
    if num_classes == 1:
        model.add(Dense(1, activation='sigmoid'))  # Use sigmoid for binary classification
        loss = 'binary_crossentropy'
    elif num_classes == 2:
        model.add(Dense(1, activation='sigmoid'))  # Use sigmoid for binary classification
        loss = 'binary_crossentropy'
    else:
        model.add(Dense(num_classes, activation='softmax'))  # Use softmax for multi-class
        loss = 'categorical_crossentropy'

    return model, loss

# Create model
model, loss_function = create_prescription_model(
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 1),
    num_classes=num_classes
)

model.compile(
    optimizer='adam',
    loss=loss_function,
    metrics=['accuracy']
)

model.summary()

# =======================================
# ğŸš€ 6. Train Model
# =======================================
if len(X_train) > 0:
    print("ğŸš€ Training model...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=5,
        batch_size=16,
        verbose=1
    )

    # Save model
    model.save("prescription_model.h5")
    print("âœ… Model saved as 'prescription_model.h5'")
else:
    print("âš ï¸ Not enough data for training. Using OCR-only approach.")

# =======================================
# ğŸ©º 7. Enhanced Prescription OCR System
# =======================================
def enhanced_prescription_ocr(image_path):
    """Enhanced OCR specifically for medical prescriptions"""

    # Read and preprocess image
    img = cv2.imread(image_path)
    if img is None:
        return "âŒ Could not read image"

    # Multiple preprocessing techniques
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Technique 1: Simple threshold
    _, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Technique 2: Adaptive threshold
    thresh2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY, 11, 2)

    # Technique 3: Noise removal + threshold
    denoised = cv2.medianBlur(gray, 3)
    _, thresh3 = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Try OCR with different preprocessings and configurations
    texts = []

    for i, processed_img in enumerate([thresh1, thresh2, thresh3]):
        # Try different PSM modes
        for psm in [6, 8, 11, 13]:
            try:
                config = f'--psm {psm} -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789-.,:/ ()'
                text = pytesseract.image_to_string(processed_img, config=config)
                if text.strip():
                    texts.append(text.strip())
            except:
                continue

    # Return the longest (most likely complete) text
    if texts:
        best_text = max(texts, key=len)
        return best_text
    else:
        return "âŒ No text detected"

# =======================================
# ğŸŒ 8. OCR.Space API Integration
# =======================================
class OCRSpaceAPI:
    def __init__(self, api_key, ocr_url):
        self.api_key = api_key
        self.ocr_url = ocr_url

    def extract_text_with_api(self, image_path):
        """Extract text using OCR.Space API"""
        try:
            print(f"ğŸ”— Calling OCR.Space API...")

            with open(image_path, 'rb') as image_file:
                # Prepare the payload
                payload = {
                    'apikey': self.api_key,
                    'language': 'eng',
                    'isOverlayRequired': False,
                    'OCREngine': 2  # Engine 2 is more accurate
                }

                # Send POST request
                response = requests.post(
                    self.ocr_url,
                    files={'filename': image_file},
                    data=payload,
                    timeout=30
                )

            # Parse response
            if response.status_code == 200:
                result = response.json()
                if 'ParsedResults' in result and len(result['ParsedResults']) > 0:
                    extracted_text = result['ParsedResults'][0]['ParsedText']
                    print("âœ… OCR.Space API call successful")
                    return extracted_text.strip()
                else:
                    print(f"âŒ API Error: {result.get('ErrorMessage', 'Unknown error')}")
                    return None
            else:
                print(f"âŒ HTTP Error: {response.status_code}")
                return None

        except Exception as e:
            print(f"âŒ API call failed: {str(e)}")
            return None

# =======================================
# ğŸ”„ 9. Text Comparison & Decision System
# =======================================
class PrescriptionAnalyzer:
    def __init__(self, api_key, ocr_url, model_path=None):
        self.ocr_api = OCRSpaceAPI(api_key, ocr_url)
        self.model = None
        if model_path and os.path.exists(model_path):
            try:
                self.model = load_model(model_path)
                print("âœ… Model loaded successfully")
            except:
                print("âš ï¸ Could not load model, using OCR-only approach")

    def compare_and_decide(self, image_path):
        """Compare model OCR vs API OCR and make decision"""
        print(f"\nğŸ” COMPARING OCR SYSTEMS:")
        print("=" * 60)

        # Get our model's OCR prediction
        our_ocr_text = enhanced_prescription_ocr(image_path)

        # Get API OCR prediction
        api_ocr_text = self.ocr_api.extract_text_with_api(image_path)

        print(f"ğŸ“„ Image: {os.path.basename(image_path)}")

        # Display both results
        print(f"\nğŸ¤– OUR MODEL OCR RESULTS:")
        print("-" * 40)
        print(our_ocr_text if our_ocr_text else "No text detected")

        print(f"\nğŸŒ OCR.SPACE API RESULTS:")
        print("-" * 40)
        print(api_ocr_text if api_ocr_text else "No text detected")

        # Compare and decide
        final_text, decision_reason = self._make_decision(our_ocr_text, api_ocr_text)

        print(f"\nğŸ¯ FINAL DECISION:")
        print("-" * 40)
        print(f"ğŸ“ Final Text: {final_text}")
        print(f"ğŸ’¡ Reason: {decision_reason}")

        return {
            'our_ocr': our_ocr_text,
            'api_ocr': api_ocr_text,
            'final_text': final_text,
            'decision_reason': decision_reason
        }

    def _make_decision(self, our_text, api_text):
        """Decide which text to use as final output"""

        # Handle cases where one or both failed
        if not our_text and not api_text:
            return "âŒ No text detected from either system", "Both OCR systems failed"

        elif not api_text:
            return our_text, "API failed - Using our model's OCR"

        elif not our_text:
            return api_text, "Our model failed - Using API OCR"

        # Check if texts are similar
        elif self._are_texts_similar(our_text, api_text):
            return api_text, "Texts are similar - Using API OCR as it's more reliable"

        else:
            # Texts are different - prefer API output
            return api_text, "Texts mismatch - Using API OCR as final output"

    def _are_texts_similar(self, text1, text2, similarity_threshold=0.7):
        """Check if two texts are similar enough"""
        if not text1 or not text2:
            return False

        # Convert to lowercase for comparison
        text1_lower = text1.lower()
        text2_lower = text2.lower()

        # Simple similarity check using SequenceMatcher
        from difflib import SequenceMatcher
        similarity = SequenceMatcher(None, text1_lower, text2_lower).ratio()

        print(f"ğŸ“Š Similarity Score: {similarity:.2f}")

        return similarity >= similarity_threshold

# =======================================
# ğŸ¥ 10. Medical Information Extraction
# =======================================
def extract_medical_info(text):
    """Just return the text without any processing"""
    print("\nEXTRACTED TEXT:")
    print(text)
    return text

# =======================================
# ğŸš€ 11. MAIN EXECUTION
# =======================================

# Initialize OCR.Space API
API_KEY = "*********************"
OCR_URL = "https://api.ocr.space/parse/image"

# Initialize analyzer
analyzer = PrescriptionAnalyzer(api_key=API_KEY, ocr_url=OCR_URL, model_path="prescription_model.h5")

# Upload prescription image
print("\nğŸ“¤ Upload a prescription image for analysis:")
uploaded_prescription = files.upload()

if uploaded_prescription:
    prescription_path = list(uploaded_prescription.keys())[0]

    print(f"\nğŸ” Analyzing prescription: {prescription_path}")

    # Display the image
    img = cv2.imread(prescription_path)
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Uploaded Prescription')
    plt.axis('off')
    plt.show()

    # Compare OCR systems and get final text
    comparison_results = analyzer.compare_and_decide(prescription_path)

    # Extract medical information from final text
    final_text = comparison_results['final_text']

    if not final_text.startswith("âŒ"):
        medical_info = extract_medical_info(final_text)

        # Display the complete extracted text
        print("\nğŸ“„ COMPLETE EXTRACTED TEXT:")
        print("=" * 60)
        print(final_text)
        print("=" * 60)

        # Save results
        results_data = {
            'image_file': prescription_path,
            'comparison_results': comparison_results,
            'medical_info': medical_info,
            'timestamp': str(np.datetime64('now'))
        }

        with open('prescription_analysis_results.json', 'w') as f:
            json.dump(results_data, f, indent=2)

        print(f"\nğŸ’¾ Results saved to 'prescription_analysis_results.json'")

    print(f"\nâœ… PRESCRIPTION ANALYSIS COMPLETED!")

else:
    print("âŒ No prescription image uploaded.")

# =======================================
# ğŸ“Š 12. PERFORMANCE METRICS
# =======================================
def show_performance_summary():
    """Show performance comparison between OCR systems"""
    print(f"\nğŸ“Š PERFORMANCE SUMMARY:")
    print("=" * 50)
    print("ğŸ”¹ Our Model OCR:")
    print("   - Pros: Free, customizable, works offline")
    print("   - Cons: May have lower accuracy on complex prescriptions")
    print("\nğŸ”¹ OCR.Space API:")
    print("   - Pros: Higher accuracy, better text recognition")
    print("   - Cons: Requires API key, internet connection")
    print("\nğŸ¯ Decision Logic:")
    print("   - If both agree: Use API output (more reliable)")
    print("   - If disagree: Use API output (higher accuracy)")
    print("   - If API fails: Fallback to our model")
    print("   - If both fail: Manual review required")

show_performance_summary()

print(f"\nğŸ‰ MEDICAL PRESCRIPTION RECOGNITION SYSTEM READY!")
print("=" * 60)
```
